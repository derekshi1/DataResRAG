# Course Webscraping

```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
major_json = "data_theory_req.json"
with open(major_json, "r") as file:
    data = json.load(file)
```

```python
def extract_classnames(obj):
    class_names = []
    if isinstance(obj, list):
        for item in obj:
            class_names.extend(extract_classnames(item))
    elif isinstance(obj, dict):
        for key, value in obj.items():
            class_names.extend(extract_classnames(value))
    elif isinstance(obj, str):
        class_names.append(obj)
    return class_names
```

```python
dt_class_names = extract_classnames(data)
dt_class_names = [name.replace(" ", "") for name in dt_class_names]
```

```python
response = requests.get(url)
if response.status_code != 200:
    print("Failed to fetch the page.")
    exit()

#print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
descriptions = soup.find('div')

# Extract and print descriptions
for idx, desc in enumerate(descriptions, start=1):
    description_text = desc.get_text(strip=True)
    print(f"Description {idx}: {description_text}")
```
